```python
import networkx as nx
import matplotlib as plt
```

```python
H = nx.triangular_lattice_graph(5,6,False, True)
nx.draw_spring(H, with_labels=True)
```

```python
H.remove_node((1,2))
```

```python
nx.draw_spring(H, with_labels=True)
```

```python
path = nx.shortest_path(H, source=(1,1), target=(2,3))
print(path)
pos = nx.spring_layout(H)
print(path)
path_edges = list(zip(path, path[1,:]))
nx.draw_networkx_nodes(H,pos,  nodelist=path, node_color='r')
nx.draw_networkx_edges(H, pos, edgelist=path_edges, edge_color='r', width=5)

plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
G = nx.karate_club_graph()
pos = nx.spring_layout(G)
nx.draw(G,pos,node_color='k')
# draw path in red
path = nx.shortest_path(G,source=14,target=16)
path_edges = list(zip(path,path[1:]))
print(path)
print(path_edges)
nx.draw_networkx_nodes(G,pos,nodelist=path,node_color='r')
nx.draw_networkx_edges(G,pos,edgelist=path_edges,edge_color='r',width=10)
plt.axis('equal')
plt.show()
```
