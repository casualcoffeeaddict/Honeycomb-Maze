```python
H = nx.triangular_lattice_graph(10,11,False, True)
nx.draw_spring(H, with_labels=True)
print(list(H.nodes))#
```

Relabel the nodes to the x + y + z = 0 plane

```python
# define remapping function
def remapping(element):
    x = element[0]
    y = element[1]
    return x, y, -(x + y)

print(remapping)
```

```python
# get list of nodes from graph
node_list = list(H.nodes)
print(node_list)
G = nx.relabel_nodes(H, remapping)
print(list(G.nodes))

# G.remove_node((3,5,-8), ())
```

```python
ax = nx.draw_spring(G, with_labels=True)
```

```python
path = nx.shortest_path(H, source=(1,1), target=(2,3))
print(path)

# path = list(sum(path, ()))
# print(path)

pos = nx.spring_layout(H)
path_edges = list(zip(path, path[1:]))
print(path_edges)

nx.draw_networkx_nodes(H,pos,  nodelist=path, node_color='r')
nx.draw_networkx_edges(H, pos, edgelist=path_edges, edge_color='r', width=5,)

plt.axis('equal')
plt.show()
```
